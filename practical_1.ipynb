```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('tested.csv')
df.head()
```

```python
print("Shape of the dataset:", df.shape)
print("\n--- Data Information ---")
df.info()
print("\n--- Missing Values ---")
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
df.drop(columns=['Cabin'], inplace=True)
print("Null values after cleaning:")
print(df.isnull().sum())
```

```python
print("--- Descriptive Statistics ---")
df.describe()
```

```python
# Plot Survival Count
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Survived', palette='Set1')
plt.title('Count of Survived (1) vs Not Survived (0)')
plt.show()

# Plot Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='teal')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.show()
```

```python
# Plot 3: Survival Rate by Sex
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Sex', y='Survived', ci=None, palette='coolwarm')
plt.title('Survival Rate: Male vs Female')
plt.show()

# Plot 4: Survival Rate by Passenger Class
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Pclass', y='Survived', ci=None, palette='magma')
plt.title('Survival Rate by Ticket Class (1st, 2nd, 3rd)')
plt.show()
```

```python
# Survival Rate by Siblings/Spouses (SibSp)
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='SibSp', y='Survived', ci=None, palette='muted')
plt.title('Survival Rate based on Siblings/Spouses on board')
plt.show()

# Survival Rate by Parents/Children (Parch)
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='Parch', y='Survived', ci=None, palette='pastel')
plt.title('Survival Rate based on Parents/Children on board')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
numeric_data = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_data.corr(), annot=True, cmap='Blues', fmt='.2f')
plt.title('Correlation Heatmap of Variables')
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='FamilySize', y='Survived', ci=None, palette='Set2')
plt.title('Survival Rate by Total Family Size')
plt.show()
```
